#### Kualitas Makanan

````python
#Kualitas Makanan Buruk
def mkn_buruk(mkn):
  nilai_mkn=0
  if (mkn<=3):
    nilai_mkn=1
  else:
    if (mkn>3) and (mkn<4):
      nilai_mkn=(-mkn+4)/(4-3)
    else:
       nilai_mkn=0
  return round(nilai_mkn,2)

#Kualitas Makanan Cukup
def mkn_cukup(mkn):
  nilai_mkn_cukup=0
  if (mkn<=3) or (mkn>=8):
    nilai_mkn_cukup=0
  else:
    if (mkn>3) and (mkn<4):
      nilai_mkn_cukup=(mkn-3)/(4-3)
    else:
      if (mkn>7) and (mkn<8):
        nilai_mkn_cukup=(-mkn+8)/(8-7)
      else:
        nilai_mkn_cukup=1
  return round(nilai_mkn_cukup,2)

#Kualitas Makanan Bagus
def mkn_bagus(mkn):
  nilai_mkn_bagus=0
  if (mkn<=7):
    nilai_mkn_bagus=0
  else:
    if (mkn>7) and (mkn<8):
      nilai_mkn_bagus=(mkn-7)/(8-7)
    else:
      nilai_mkn_bagus=1
  return round(nilai_mkn_bagus,2)
````

## Grafik Keanggotaan

````python
def representasi_linear_turun(x, start, end):
    return np.piecewise(x, [x <= start, (x > start) & (x < end), x >= end], [1, lambda x: (end - x) / (end - start), 0])

def representasi_linear_naik(x, start, end):
    return np.piecewise(x, [x <= start, (x > start) & (x < end), x >= end], [0, lambda x: (x - start) / (end - start), 1])

def representasi_segitiga(x, start, mid, end):
    return np.piecewise(x, [(x >= start) & (x < mid), x == mid, (x > mid) & (x <= end)], [lambda x: (x - start) / (mid - start), 1, lambda x: (end - x) / (end - mid)])

def representasi_trapesium(x, start, mid_left, mid_right, end):
    return np.piecewise(x, [(x >= start) & (x < mid_left), (x >= mid_left) & (x <= mid_right), (x > mid_right) & (x <= end)], [lambda x: (x - start) / (mid_left - start), 1, lambda x: (end - x) / (end - mid_right)])

def print_header(text, width=30, char='-'):
    centered_text = text.center(width)
    horizontal_line = char * width
    print(horizontal_line)
    print(centered_text)
    print(horizontal_line)
````

### Fungsi Keanggotaan Kualitas Makanan

````python
def fungsi_keanggotaan_kualitas(x):
    buruk = representasi_linear_turun(x,3,4)
    cukup = representasi_trapesium(x,0,4,7,8)
    bagus = representasi_linear_naik(x, 7, 8)
    print(f"Input: {x}, Buruk: {buruk}, Cukup: {cukup}, Bagus: {bagus}")
    return buruk, cukup, bagus
````

#### Plot Grafik Fungsi Keanggotaan Kualitas Makanan

````python
def plot_membership_functions(x, y, labels, title):
    plt.figure(figsize=(9, 4))
    plt.xticks(x)
    for i in range(len(y)):
        plt.plot(x, y[i], label=labels[i])
        differences = np.diff(y[i])

        # Temukan indeks di mana terjadi perubahan tanda
        turning_point_indices = np.where(np.diff(np.sign(differences)))[0] + 1
        turning_point_indices = turning_point_indices[y[i][turning_point_indices] == 1]

        for j in turning_point_indices:
            plt.vlines(x[j], 0, y[i][j], linestyle="dashed", color="gray", alpha=0.7)

    plt.title(title)
    plt.xlabel("Nilai")
    plt.ylabel("Derajat Keanggotaan")
    plt.legend()
    plt.show()

# Data input
x_kualitas = np.arange(0, 11, 1)
y_kualitas = fungsi_keanggotaan_kualitas(x_kualitas)

# Plot grafik
plot_membership_functions(x_kualitas, y_kualitas, ["Buruk", "Cukup", "Bagus"], "Fungsi Keanggotaan Kualitas Makanan")
````

### Fungsi Keanggotaan Kondisi Tempat

````python
def fungsi_keanggotaan_kondisi(x):
    tidak_nyaman = representasi_linear_turun(x,3,5)
    cukup = representasi_segitiga(x,3,5,8)
    nyaman = representasi_linear_naik(x, 5, 8)

    return tidak_nyaman, cukup, nyaman
````

## Fuzzifikasi

````python
def fuzzifikasi(mkn,pelayanan,kondisi):
    buruk_k, cukup_k, bagus_k = fungsi_keanggotaan_kualitas(mkn)
    buruk_p, cukup_p, baik, sgt_baik = fungsi_keanggotaan_pelayanan(pelayanan)
    tidak_nyaman, cukup_c, nyaman = fungsi_keanggotaan_kondisi(kondisi)

  #Menghitung nilai keanggotaan untuk setiap kategori
    nilai_keanggotaan_kualitas = {
        'Buruk': buruk_k,
        'Cukup': cukup_k,
        'Bagus': bagus_k
  }

    nilai_keanggotaan_pelayanan = {
        'Buruk': buruk_p,
        'Cukup': cukup_p,
        'Baik': baik,
        'Sangat Baik' : sgt_baik
    }

    nilai_keanggotaan_kondisi = {
        'Tidak Nyaman': tidak_nyaman,
        'Cukup Nyaman': cukup_c,
        'Nyaman': nyaman
    }

    nilai_keanggotaan_kualitas_tidak_nol = {key: value for key, value in nilai_keanggotaan_kualitas.items() if not np.all(value == 0)}
    nilai_keanggotaan_kelembapan_tidak_nol = {key: value for key, value in nilai_keanggotaan_pelayanan.items() if not np.all(value == 0)}
    nilai_keanggotaan_kondisi_tidak_nol = {key: value for key, value in nilai_keanggotaan_kondisi.items() if not np.all(value == 0)}

    return nilai_keanggotaan_kualitas_tidak_nol, nilai_keanggotaan_kelembapan_tidak_nol, nilai_keanggotaan_kondisi_tidak_nol
````

# **Contoh Penggunaan Fuzzifikasi** :

````python
input_kualitas = float(input("Masukan kualitas : "))
input_pelayanan = float(input("Masukan pelayanan : "))
input_kondisi = float(input("Masukan kondisi : "))

nilai_keanggotaan_kualitas, nilai_keanggotaan_pelayanan, nilai_keanggotaan_kondisi = fuzzifikasi(input_kualitas, input_pelayanan, input_kondisi)

print_header("Hasil Fuzzifikasi")
print("\n------<Nilai Fuzzifikasi Kualitas>------")
# Print nilai keanggotaan untuk setiap kategori
for kategori, nilai in nilai_keanggotaan_kualitas.items():
    print(f'- {kategori}: {nilai:.2f}')

print("\n------<Nilai Fuzzy Pelayanan>------")
for kategori, nilai in nilai_keanggotaan_pelayanan.items():
    print(f'- {kategori}: {nilai:.2f}')

print("\n------<Nilai Fuzzy Kondisi Tempat>------")
for kategori, nilai in nilai_keanggotaan_kondisi.items():
    print(f'- {kategori}: {nilai:.2f}')
````

## Inferensi

### Menentukan Rules

````python
def inferensi(mkn,pelayanan,kondisi):
    rules = {
        'Buruk': {
            'Buruk' :{
                'Tidak Nyaman': 'Kecewa Berat',
                'Cukup Nyaman': 'Kecewa Berat',
                'Nyaman': 'Kecewa Berat'
            },
            'Cukup' : {
                'Tidak Nyaman': 'Kecewa Berat',
                'Cukup Nyaman': 'Kecewa Berat',
                'Nyaman': 'Kecewa Berat'
            },
            'Baik' : {
                'Tidak Nyaman': 'Kecewa Berat',
                'Cukup Nyaman': 'Kecewa Berat',
                'Nyaman': 'Cukup Puas'
        },
            'Sangat Baik': {
                'Tidak Nyaman': 'Kecewa Berat',
                'Cukup Nyaman': 'Cukup Puas',
                'Nyaman': 'Cukup Puas'
            }
      },
        'Cukup': {
            'Buruk' : {
                'Tidak Nyaman': 'Kecewa Berat',
                'Cukup Nyaman': 'Kecewa Berat',
                'Nyaman': 'Cukup Puas'
            },
            'Cukup' : {
                'Tidak Nyaman': 'Cukup Puas',
                'Cukup Nyaman': 'Cukup Puas',
                'Nyaman': 'Puas'
            },
            'Baik' : {
                'Tidak Nyaman': 'Cukup Puas',
                'Cukup Nyaman': 'Puas',
                'Nyaman': 'Puas'
            },
            'Sangat Baik': {
                'Tidak Nyaman': 'Cukup Puas',
                'Cukup Nyaman': 'Puas',
                'Nyaman': 'Cukup Puas'
            }
      },
        'Bagus':{
             'Buruk' : {
                'Tidak Nyaman': 'Cukup Puas',
                'Cukup Nyaman': 'Cukup Puas',
                'Nyaman': 'Puas'
            },
            'Cukup' : {
                'Tidak Nyaman': 'Cukup Puas',
                'Cukup Nyaman': 'Puas',
                'Nyaman': 'Puas'
            },
            'Baik' : {
                'Tidak Nyaman': 'Puas',
                'Cukup Nyaman': 'Puas',
                'Nyaman': 'Puas'
            },
            'Sangat Baik': {
                'Tidak Nyaman': 'Puas',
                'Cukup Nyaman': 'Puas',
                'Nyaman': 'Puas'
            }
      }
    }
    df_rules = pd.DataFrame.from_dict(rules, orient='index').transpose()
    rules_digunakan = {}

    print_header("Rules yang digunakan dan implikasi MIN", width=45)
    for kategori_kualitas, nilai_kualitas in mkn.items():
      for kategori_pelayanan, nilai_pelayanan in pelayanan.items():
        for kategori_kondisi, nilai_kondisi in kondisi.items():
          durasi = rules[kategori_kualitas][kategori_pelayanan][kategori_kondisi]
          if durasi not in rules_digunakan:
              rules_digunakan[durasi] = []
          implikasi_min = np.round(np.fmin(nilai_kualitas, np.fmin(nilai_pelayanan, nilai_kondisi)), 2)
          rules_digunakan[durasi].append(implikasi_min)
          print(f'{kategori_kualitas}({nilai_kualitas:.2f})  {kategori_pelayanan}({nilai_pelayanan:.2f})  {kategori_kondisi}({nilai_kondisi:.2f}) = {durasi}({implikasi_min})')
    display(df_rules)
    return rules_digunakan

implikasi_min = inferensi(nilai_keanggotaan_kualitas, nilai_keanggotaan_pelayanan,nilai_keanggotaan_kondisi)
implikasi_max = {}
print_header("Implikasi MAX", width=35)
for durasi, nilai in implikasi_min.items():
    implikasi_max[durasi] = np.max(nilai)
    print(f'{durasi} = MAX{nilai} = {np.max(nilai)}')
````

### Mamdani

````python
def deffuzifikasi(x, y, implikasi, x_sampel):
    kecewa_berat,cukup,puas = y
    nilai_keanggotaan_kepuasan = {
        'Kecewa Berat': kecewa_berat,
        'Cukup Puas': cukup,
        'Puas': puas
    }

    plt.figure(figsize=(9, 4))
    plt.xticks(np.arange(0, 11, 1))

    for kepuasan, nilai in nilai_keanggotaan_kepuasan.items():
        plt.plot(x, nilai, label = kepuasan)

        differences = np.diff(nilai)
        # Temukan indeks di mana terjadi perubahan tanda
        turning_point_indices = np.where(np.diff(np.sign(differences)))[0] + 1
        turning_point_indices = turning_point_indices[nilai[turning_point_indices] == 1]

        for j in turning_point_indices:
            plt.vlines(x[j], 0, nilai[j], linestyle="dashed")


    # grafik arsir gabungan dari inferensi sebelumnya
    y_union = np.zeros_like(x, dtype=float) # Initialize y_union with zeros
    for kepuasan, nilai in implikasi.items():
        clipped_values = np.clip(nilai_keanggotaan_kepuasan[kepuasan], 0, nilai)
        y_union = np.maximum(y_union, clipped_values) # Update with maximum

    y_sampel = y_union[np.isin(np.round(x, 1), x_sampel)]


    plt.scatter(x_sampel, y_sampel, color='red', marker='o', label='Sampel')
    plt.hlines(y_sampel, min(x), x_sampel, linestyle="dashed", color='black')
    plt.fill_between(x, y_union, alpha=0.3, color='gray', hatch='//')

    # Mengarsir area irisan di antara kedua kurva
    plt.title("Tingkat Kepuasan Pelanggan")
    plt.xlabel("Skala Kepuasan Pelanggan")
    plt.ylabel("Derajat Keanggotaan")
    # plt.legend()
    plt.show()
    print(f"{np.sum(x_sampel * y_sampel):.2f} / {np.sum(y_sampel):.2f}")
    return np.sum(x_sampel * y_sampel) / np.sum(y_sampel)

def fungsi_keanggotaan_kepuasan(x):
    kecewa_berat = representasi_linear_turun(x, 3, 4)
    cukup = representasi_trapesium(x, 3, 4, 7, 8)
    puas = representasi_linear_naik(x, 7, 8)

    return kecewa_berat, cukup, puas

x_kepuasan = np.arange(0,11, 1, dtype=float)
y_kepuasan = fungsi_keanggotaan_kepuasan(x_kepuasan)

titik_sampel = [4,5,6,7,8,9,10]
hasil_akhir = deffuzifikasi(x_kepuasan, y_kepuasan , implikasi_max, titik_sampel)

print(f"Nilai Kelayakan : {hasil_akhir:.2f}")
````

### Sugeno

````python
def sugeno(x, y, implikasi, x_sampel):
    kecewa_berat,cukup,puas = y
    nilai_keanggotaan_kepuasan = {
        'Kecewa Berat': kecewa_berat,
        'Cukup Puas': cukup,
        'Puas': puas
    }
    colors = ['#448dc0', '#ff9436', '#4fb04f']
    plt.figure(figsize=(9, 4))
    plt.xticks(np.arange(0, 11, 1))
    plt.xlim(0, 11)

    # grafik arsir gabungan dari inferensi sebelumnya
    y_union = np.zeros_like(x, dtype=float) # Initialize y_union with zeros
    for kepuasan, nilai in implikasi.items():
        clipped_values = np.clip(nilai_keanggotaan_kepuasan[kepuasan], 0, nilai)
        y_union = np.maximum(y_union, clipped_values) # Update with maximum

    x_sampel = [val for val in x_sampel if min(x) <= val <= max(x)]
    y_sampel = y_union[np.isin(np.round(x, 1), x_sampel)]

    plt.vlines(x_sampel, 0, 1, linestyle="dashed", color='black', alpha=0.3)
    for i, x_val in enumerate(x_sampel):
          color = colors[i % len(colors)]  # Mengambil warna dari daftar secara bergiliran
          if i < len(y_sampel):
            plt.vlines(x_val, 0, y_sampel[i], linestyle="solid", color=color)
    for x_val, y_val in zip(x_sampel, y_sampel):
        plt.hlines(y_val, min(x), x_val, linestyle="dashed", color='black', alpha=0.3)

    plt.scatter(x_sampel, y_sampel, color='red', marker='o', label='Sampel')

    # Mengarsir area irisan di antara kedua kurva
    plt.title("Tingkat Kepuasan Pelanggan")
    plt.xlabel("Skala Kepuasan Pelanggan")
    plt.ylabel("Derajat Keanggotaan")
    # plt.legend()
    plt.show()
    print(f"{np.sum(x_sampel * y_sampel):.2f} / {np.sum(y_sampel):.2f}")
    return np.sum(x_sampel * y_sampel) / np.sum(y_sampel)

def sampel(implikasi_max):
    sampel = []
    range_fungsi_linguistik = {
        'Kecewa Berat': [3],
        'Cukup Puas': [4, 7],
        'Puas': [8]
    }
    for nama_fungsi in implikasi_max.keys():
      if nama_fungsi == 'Rendah':
        sampel.append(int(range_fungsi_linguistik[nama_fungsi][0]))
      else :
        sampel.append(range_fungsi_linguistik[nama_fungsi][0])

    sampel.sort()
    return sampel

x_kepuasan = np.arange(1, 11, 1, dtype=float)
y_kepuasan = fungsi_keanggotaan_kepuasan(x_kepuasan)

titik_sampel = sampel(implikasi_max)
hasil_akhir = sugeno(x_kepuasan, y_kepuasan, implikasi_max, titik_sampel)

print(f"Nilai Kepuasan : {hasil_akhir:.2f}")
````
